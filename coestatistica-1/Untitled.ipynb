```python
import json
import io
```

```python
json_data = {}
```

```python
with io.open('submission.json', 'r') as input_file:
    json_data = json.load(input_file)
```

```python
# {
#   "PR": {
#     "moda": 850,
#     "mediana": 650,
#     "media": 648.9612940497,
#     "desvio_padrao": 98.6071859131
#   },
#   "RS": {
#     "moda": 850,
#     "mediana": 650,
#     "media": 651.1051428571,
#     "desvio_padrao": 95.1365984138
#   },
#   "SC": {
#     "moda": 850,
#     "mediana": 653,
#     "media": 649.5376527423,
#     "desvio_padrao": 97.2334927934
#   }
# }
filtered_df[filtered_df['estado_residencia'] == 'SC'].mode()['pontuacao_credito'][0]
```

```python
json_data['SC']['moda'] = int(filtered_df[filtered_df['estado_residencia'] == 'SC'].mode()['pontuacao_credito'][0])
json_data['SC']['mediana'] = int(estado_described[('pontuacao_credito', '50%')]['SC'])
json_data['SC']['media'] = estado_described[('pontuacao_credito', 'mean')]['SC']
json_data['SC']['desvio_padrao'] = estado_described[('pontuacao_credito', 'std')]['SC']
```

```python
json_data['PR']['moda'] = int(filtered_df[filtered_df['estado_residencia'] == 'PR'].mode()['pontuacao_credito'][0])
json_data['PR']['mediana'] = int(estado_described[('pontuacao_credito', '50%')]['PR'])
json_data['PR']['media'] = estado_described[('pontuacao_credito', 'mean')]['PR']
json_data['PR']['desvio_padrao'] = estado_described[('pontuacao_credito', 'std')]['PR']
```

```python
json_data['RS']['moda'] = int(filtered_df[filtered_df['estado_residencia'] == 'RS'].mode()['pontuacao_credito'][0])
json_data['RS']['mediana'] = int(estado_described[('pontuacao_credito', '50%')]['RS'])
json_data['RS']['media'] = estado_described[('pontuacao_credito', 'mean')]['RS']
json_data['RS']['desvio_padrao'] = estado_described[('pontuacao_credito', 'std')]['RS']
```

```python
with io.open('submission.json', 'w') as out_file:
    json.dump(json_data, out_file)
```
